# This is function to calculate the exponential using 

# the exponential series that takes the number of terms of

# the series as an argument, use default of 100


## Defining myexpo function 

```python
def myexpo(x,n=100,debug=False):
    import numpy as np
    sum=0
    flag=0
    tol=1.0e-14
    for i in range(n+1):
        t2=sum
        sum=sum+(np.power(x,i)/np.math.factorial(i))
        t1=sum
        if debug:
            if(abs(t2-t1))<tol:
                print(f"Series is converged at i={i}")
                break
               
            
    return sum
```

#### testing of convergence

```python
x1=-5
myexpo(x1,100,True)
```

## Making 1000 points x between 0 to 100 and then calling myexpo and numpy exp function 

```python
import numpy as np
x=np.linspace(0,100,1000)
y=myexpo(x)
z=np.exp(x)
```

##  Excution time comparision of both function

```python
print("Excution time taken by myexpo function is:")
%timeit y
print("Excution time taken by numpy exp function is:")
%timeit z
```

### Ploting myexpo and numpy exp function to compare results

```python
import matplotlib.pyplot as plt
plt.plot(x, y, 'o-',  label='Using myexponential function')
plt.plot(x, z, 'r--', label='Using numpy exp function')
plt.xlabel("x")
plt.ylabel("y & z")
plt.legend()
plt.title('Myexponential vs numpy exp')
plt.show()
```
